```python
import numpy as np
```

## Highest Marks per Student

```python
marks = np.array([
    [1, 78, 85, 88, 90],
    [2, 65, 70, 72, 68],
    [3, 90, 92, 89, 95],
    [4, 55, 60, 58, 62],
    [5, 80, 78, 85, 88]
])
print("Highest marks of each student: ", np.max(marks[:,1:],axis= 1))
```

## Average Marks per student

```python

print("Average Marks of students: ",np.average(marks[:,1:] , axis = 1) )
```

## Finding topper 

```python
total = np.sum(marks[:,1:], axis = 1)
print(total)
print("\n")


highest_marks = np.argmax(total)

print("Topper ID: ", highest_marks)
print("\n")
topper_data = marks[2]
print(topper_data[1:])

```
