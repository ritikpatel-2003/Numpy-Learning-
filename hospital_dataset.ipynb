```python

import numpy as np

# [PatientID, Age, HeartRate, BP, SugarLevel]
patients = np.array([
    [1, 25, 72, 120, 90],
    [2, 40, 85, 140, 110],
    [3, 60, 78, 130, 150],
    [4, 35, 90, 150, 95],
    [5, 50, 88, 145, 130]
])
```

## Find minimum age

```python
min_index = np.argmin(patients[:,1])
print(min_index)
print("Minimum Age:", patients[min_index,1])
print("Patient Details:", patients[min_index])
```

## Find Maximum Age 

```python
max_age = np.argmax(patients[:,1])
print(max_age)

# Printing the age of the Patients

print("Maximum age of the patients is: ", patients[max_age,1])

print("Details of the Patient: ", patients[max_age])
```

## Find all patients where: BP > 140 OR Sugar > 140

```python

import numpy as np

# [PatientID, Age, HeartRate, BP, SugarLevel]
patients = np.array([
    [1, 25, 72, 120, 90],
    [2, 40, 85, 140, 110],
    [3, 60, 78, 130, 150],
    [4, 35, 90, 150, 95],
    [5, 50, 88, 145, 130]
])

critical = patients[(patients[:,3]>140) | (patients[:,4]>140)]


print("Critical patients: ", critical)
print(" No of Critical Patients: ", len(critical))

```

##  Patients with Normal Sugar

```python
normal_sugar = patients[patients[:,4]<100]
print("Patients with normal sugar: ",normal_sugar)
```

##  Average Health Score

```python
hr  = patients[:,2]
print(hr)
bp = patients[:,3]
print(bp)
sugar = patients[:,4]
print(sugar)

average_health = (hr+bp+sugar)/3
print(average_health)
```
