```python
import numpy as np 

# [OrderID, CustomerAge, ItemsCount, TotalAmount, Discount]
orders = np.array([
    [101, 22, 2, 500, 50],
    [102, 35, 5, 1500, 200],
    [103, 28, 1, 300, 20],
    [104, 45, 3, 1200, 150],
    [105, 32, 4, 2000, 300],
    [106, 26, 2, 700, 70],
    [107, 38, 6, 2500, 400],
    [108, 30, 3, 1000, 100]
])
```

## Print all Total Amount values

```python
print("Total Amount Values: ", orders[:,3])
```

## Print all Discount values

```python
print("All the Discounts Values: ", orders[:,4])
```

## Print first 4 orders

```python
print("First 4 Orders",orders[:4:])
```

## Find minimum and maximum Total Amount

```python
# Maximum Amount

amount_column = orders[:,3]

max_amount = np.argmax(amount_column)

print(" Customer with Maximum Amount: ",orders[max_amount])


# Minimum Amount 

min_amount = np.argmin(amount_column)

print(" Customer with Minimum Amount: ",orders[min_amount])
```

## Find order with minimum discount

```python
discount_column = orders[:,4]
min_discount = np.argmin(discount_column)
print("Customer With minimum amount: ", orders[min_discount])
```

## Find average discount

```python
discount_column = orders[:,4]
print("Average discount Per Customer: ", np.mean(discount_column))
```

## Print orders where Total Amount > 1000

```python
print("Customer having amount more than 1000:", orders[orders[:,3]>1000])
```

## Customer Age between 25 and 35

```python
print("Customer Age between 25 and 35:", orders[(orders[:,1]>=25) & (orders[:,1]<=35)])
```

##  Problem: Top 2 Highest Discount Orders

```python
discount_column = orders[:,4]
print(discount_column)

top_2 = np.argsort(discount_column)
top_2_idx = top_2[-2:]
print(top_2_idx)

top_2_order = orders[top_2_idx]
print(top_2_order)

```

## Top 3 Orders (Only ID + Amount)

```python
amount_id= orders[:,3]
print(amount_id)

# sorting the array
sorted_column = np.argsort(amount_id)
print(sorted_column)

top_3 = sorted_column[-3:]

print("Top 3 orders with id and amount : ", (orders[top_3][:,[0,3]]))



```
